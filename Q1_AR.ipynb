```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer, PorterStemmer
from nltk.tokenize import sent_tokenize , word_tokenize
import glob
import re
import os
import numpy as np
import sys
import math
from pathlib import Path
from collections import Counter
import numpy as nap
Stopwords = set(stopwords.words('english'))
ps=PorterStemmer()
import pickle
from num2words import num2words
```

```python
file_folder = 'english-corpora/*'
ps = PorterStemmer()
all_words = []
doc_words=[]
doc_word_count=[]
dict_global = {}
dict_global_test={}
Ld={}
files_with_index = {}
i=0
idx = 0
for file in glob.glob(file_folder):
    i=i+1
    print(i,"----",file)
    fname = file
    file = open(file , "r",encoding='UTF-8')  #open file
    text = file.read()
    text = re.sub(re.compile('[^a-zA-Z0-9\s]'),'',text)
    text = re.sub(re.compile('\d'),'',text)   #removing all characters except the alphabets and spaces
    tokens = word_tokenize(text)              #tockenize each file
    Ld[idx] = len(tokens)                     #storing the number of tokens
    words=[]
    for word in tokens:
        if len(word)>1:                       #stemming and removing stopwords from the tokens
            word=ps.stem(word)
            if word not in Stopwords:
                words.append(word)
    doc_words.append(words)                   #append the words of each file to a list
    counter = dict(Counter(words))            #Counting number of occurances of each word
    dict_global.update(counter)               
    doc_word_count.append(counter)
    #for w in words:
        #dict_global_test[w] = dict_global_test.get(w,0) + 1
    files_with_index[idx] = os.path.basename(fname) #store index and file name 
    idx = idx + 1
    
unique_words_all = set(dict_global.keys())       #store unique words from all files 
```

```python
tf = {x: {} for x in unique_words_all}  #term frequency : file and frequency dictionary for each word
df = {x:0 for x in unique_words_all}    #document frequency for each word
```

```python
#calculating the term and document frequencies
idx =0
for doc in doc_word_count:
    for i in doc.keys():
        df[i]=df[i]+1
        tf[i][idx]=doc[i]   
    idx=idx+1 
print("Finished")
```

```python
#calculating norm for each document

doc_norm={}
idx=0
files_count = len(files_with_index)
for i in doc_word_count:
    l2=0
    for j in i.keys():
        l2+=(i[j]*math.log(files_count/df[j]))**2
    doc_norm[idx]=(math.sqrt(l2))
    idx +=1
    print(idx)
```
